## Description
This program demonstrates the use of Python loops and conditional logic for basic number operations.

### Tasks Performed
1. Print numbers between **1 and 500** that are:
   - Not divisible by 2
   - Not divisible by 3

2. Print all **prime numbers** between 1 and 500.

---

## Concepts Used
- `for` loops
- Conditional statements (`if`)
- Modulus operator (`%`)
- Nested loops
- Basic algorithmic thinking


```python
n = 500
not_divisible = []
divisible_by_2 = []

for i in range(1, n + 1):
    if i % 2 != 0 and i % 3 != 0:
        not_divisible.append(i)
    elif i % 2 == 0:
        divisible_by_2.append(i)

print("Numbers not divisible by 2 and 3:")
print(not_divisible)

print("\nNumbers divisible by 2:")
print(divisible_by_2)
```

```python
# Part 2: Prime numbers between 1 and 500
print("Prime numbers between 1 and 500:\n")

for num in range(2, 501):
    is_prime = True
    
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            is_prime = False
            break
    
    if is_prime:
        print(num, end=" ")
```
